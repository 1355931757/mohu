## Launch WebUI
## 启动 WebUI

```
# ==================== GPT-SoVITS WebUI + ngrok（最终修复版）====================
import os, time
from IPython.display import display, Javascript
from pyngrok import ngrok, conf

# 保活
def keep_alive():
    display(Javascript('''
        function ClickConnect(){
            console.log("Colab 保活运行中...");
            document.querySelector("colab-connect-button").shadowRoot.querySelector("#connect").click();
        }
        setInterval(ClickConnect, 60000);
    '''))
    print("✅ 保活已启用")

keep_alive()

# 确保 pyngrok
!pip install -q pyngrok

# ====================== Token 配置 ======================
token = "填入你的token"  # 默认值

# 如果 token 还是占位符，则让用户输入
if token == "填入你的token" or not token:
    token = input("新 Token（回车跳过）: ").strip()

# 如果有有效 token，则配置 ngrok
if token and token != "填入你的token":
    conf.get_default().auth_token = token

# ====================== 启动 WebUI（关键修复）====================
os.environ['PATH'] = '/content/miniconda/bin:' + os.environ.get('PATH', '')
%cd /content/GPT-SoVITS

print("🚀 重新启动 WebUI（强制 0.0.0.0 + share=True）...")

# 先杀掉旧进程
!pkill -f "webui.py" || true

!source /content/miniconda/bin/activate GPTSoVITS && \
nohup python webui.py --language zh --share > webui.log 2>&1 &

print("⏳ 等待启动（50-70秒，请耐心）...")
time.sleep(60)

# ====================== ngrok 穿透 ======================
print("🔗 创建 ngrok 隧道（端口 9874）...")
ngrok.kill()

public_url = ngrok.connect(9874, "http")
print("\n🎉 公网访问地址：")
print(public_url)

print("\n📄 最新 40 行日志：")
!tail -n 40 webui.log
```
