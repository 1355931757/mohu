## 启动 WebUI

```
# ==================== GPT-SoVITS WebUI + ngrok 完整启动 ====================
import os, time
from IPython.display import display, Javascript
from pyngrok import ngrok, conf

# ====================== 1. 保活脚本 ======================
def keep_alive():
    display(Javascript('''
        function ClickConnect(){
            console.log("Colab 保活运行中...");
            document.querySelector("colab-connect-button").shadowRoot.querySelector("#connect").click();
        }
        setInterval(ClickConnect, 60000);
    '''))
    print("✅ Colab 保活已启用（每60秒自动点击）")

keep_alive()

# ====================== 2. 安装 pyngrok ======================
print("📦 正在安装 pyngrok...")
!pip install -q pyngrok

# ====================== 3. 配置 ngrok Token ======================
print("🔑 配置 ngrok Token...")

# 方法一：推荐使用 Colab Secrets（左侧钥匙图标）
try:
    from google.colab import userdata
    token = userdata.get('NGROK_TOKEN')
    if token:
        conf.get_default().auth_token = token
        print("✅ 已从 Colab Secrets 读取 NGROK_TOKEN")
    else:
        raise Exception("Token 为空")
except:
    # 方法二：手动输入（明文，不推荐长期使用）
    token = input("请输入你的 ngrok authtoken: ")
    if token.strip():
        conf.get_default().auth_token = token.strip()
        print("✅ Token 已手动设置")
    else:
        print("❌ Token 未输入，请先获取 ngrok token")

# ====================== 4. 启动 WebUI 到后台 ======================
os.environ['PATH'] = '/content/miniconda/bin:' + os.environ.get('PATH', '')
%cd /content/GPT-SoVITS

print("🚀 正在后台启动 GPT-SoVITS WebUI...")

!source /content/miniconda/bin/activate GPTSoVITS && \
nohup python webui.py --language zh > webui.log 2>&1 &

# 等待服务启动
print("⏳ 等待 WebUI 启动（建议等待 30-50 秒）...")
time.sleep(40)

# ====================== 5. ngrok 穿透 ======================
print("🔗 创建 ngrok 隧道（端口 9870）...")
ngrok.kill()  # 清理旧隧道

public_url = ngrok.connect(9870, "http")
print("\n🎉 成功！你的公网访问地址是：")
print(public_url)

# 显示实时日志
print("\n📄 最近 30 行日志：")
!tail -n 30 webui.log
```
