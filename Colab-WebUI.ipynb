```
# ==================== GPT-SoVITS 一键安装（修复 ToS 版）====================
import os
from IPython.display import display, Javascript

# 保活脚本
def keep_alive():
    display(Javascript('''
        function ClickConnect(){
            console.log("Colab 保活运行中...");
            document.querySelector("colab-connect-button").shadowRoot.querySelector("#connect").click();
        }
        setInterval(ClickConnect, 60000);
    '''))
    print("✅ 保活脚本已启用")

keep_alive()

print("🚀 开始修复并安装 GPT-SoVITS...")

# 1. 安装 Miniconda
!wget -q https://repo.anaconda.com/miniconda/Miniconda3-latest-Linux-x86_64.sh -O miniconda.sh
!bash miniconda.sh -b -p /content/miniconda
!rm -f miniconda.sh

# 添加 conda 到 PATH
os.environ['PATH'] = '/content/miniconda/bin:' + os.environ['PATH']

# 2. 接受 Anaconda ToS
print("📜 接受 Anaconda 服务条款...")
!conda tos accept --override-channels --channel https://repo.anaconda.com/pkgs/main
!conda tos accept --override-channels --channel https://repo.anaconda.com/pkgs/r

# 3. 创建 conda 环境
print("🛠️ 创建 GPTSoVITS 环境...")
!conda create -n GPTSoVITS python=3.10 -y

# 4. 克隆仓库（如果已存在则跳过）
if not os.path.exists("/content/GPT-SoVITS"):
    !git clone https://github.com/RVC-Boss/GPT-SoVITS.git /content/GPT-SoVITS
else:
    print("✅ GPT-SoVITS 仓库已存在，跳过克隆")

# 5. 进入目录并安装依赖
%cd /content/GPT-SoVITS

print("📦 正在执行 install.sh（可能需要 8-15 分钟）...")
!source /content/miniconda/bin/activate GPTSoVITS && \
pip install ipykernel && \
bash install.sh --device CU126 --source HF --download-uvr5

print("\n🎉 安装完成！")

# 验证环境
!source /content/miniconda/bin/activate GPTSoVITS && conda env list
!source /content/miniconda/bin/activate GPTSoVITS && python --version
```
